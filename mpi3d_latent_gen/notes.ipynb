# **ELBO**
$$𝓛_{𝜃, 𝜙}=𝔼_{𝑞_{𝜙}(𝐳\,|\,𝐱)}[log\,p_𝜃(\,𝐱, 𝐳\,)-log\,q_𝜙(\,𝐳\,|\,𝐱\,)]
$$

$$𝓛_{𝜃, 𝜙}=𝔼_{𝑝(𝜖)}[log\,p_𝜃(\,𝐱, 𝐳\,)-log\,q_𝜙(\,𝐳\,|\,𝐱\,)]
$$

**given that:**
$$𝑝(𝐱, 𝐳) = p(𝐱\,|\,𝐳)\,p(𝐳)$$
\
**then:**
$$𝓛_{𝜃, 𝜙}= 𝔼_{𝑝(𝜖)}[log\,(p_𝜃(𝐱\,|\,𝐳)\,p_𝜃(𝐳))-log\,q_𝜙(\,𝐳\,|\,𝐱\,)]
$$
\
\
**given that:**
$$log\,x\,y = log\,x + log\,y$$
\
**then (final version of the ELBO):**
$$
𝓛_{𝜃, 𝜙}=
𝔼_{𝑝(𝜖)}[log\,p_𝜃(𝐱\,|\,𝐳)+log\,p_𝜃(𝐳)-log\,q_𝜙(\,𝐳\,|\,𝐱\,)]
$$

$$q_𝜙(\,𝐳\,|\,𝐱\,) = q_𝜙({𝐳\,|\,𝑒_{𝜙}(𝐱)})$$

$$(𝜇, \log{𝜎^2}) = 𝑒_{𝜙}(𝐱)$$


# **Reparameterization Trick**
$$𝜖 = 𝒩(0, 𝐼)$$

$$𝒛 = 𝜇 + 𝜎 · 𝜖$$

**note that:**
$$𝜖 = \frac{𝒛 - 𝜇}{𝜎}$$


# **Gaussian Distribution**
$$𝒩(𝑥; 𝜇, 𝜎^2) = \sqrt{\frac{1}{2πσ^2}}\exp({-\frac{1}{2σ^2}(𝑥 - 𝜇)^2})$$

$$\log{𝒩(𝑥; 𝜇, 𝜎^2)} = -\frac{1}{2}{\log{2πσ^2}} - \frac{(𝑥 - 𝜇)^2}{2σ^2}$$

$$\log{𝒩(𝑥; 𝜇, 𝜎^2)} = -\frac{1}{2}({\log{2πσ^2}} + \frac{(𝑥 - 𝜇)^2}{σ^2})$$

$$\log{𝒩(𝑥; 𝜇, 𝜎^2)} = -\frac{1}{2}({\log{2π}+\log{σ^2}}+\frac{(𝑥 - 𝜇)^2}{σ^2})$$

# **Reparameterization Trick observations**
$$𝜖 = \frac{𝒛 - 𝜇}{𝜎}$$

$$𝜖^2 = \frac{(𝒛 - 𝜇)^2}{𝜎^2}$$

# **Posterior (Distribution parameters given by the encoder)**
$$𝐳 \sim q_𝜙({𝐳\,|\,𝐱}) = q_𝜙({𝐳\,|\,𝑒_{𝜙}(𝐱)})$$

$$\log{𝒩(𝑧; 𝜇, 𝜎^2)} = -\frac{1}{2}({\log{2πσ^2}}+𝜖^2)$$

$$\log{𝒩(𝑧; 𝜇, 𝜎^2)} = -\frac{1}{2}({\log{2π}} + \log{σ^2} + 𝜖^2)$$

$$\log{𝒩(𝑧; 𝜇, 𝜎^2)} = \log{q_𝜙({𝐳\,|\,𝐱})}$$

# **Prior (Standard normal prior - mean: 0, variance: 1 - initial state)**
$$𝑝_{𝜃}(𝐳) = 𝒩(𝑧; 𝜇, 𝜎^2) = 𝒩(𝑧; 0, 𝐼)$$

$$\log{𝑝_{𝜃}(𝐳)} = -\frac{1}{2}(\log{2π} + 𝑧^2)$$

# **Likelihood**
$$𝑝_𝜃(𝐱 \mid 𝐳) = 𝑝_𝜃({𝐱\,|\,𝑑_{𝜃}(𝐳)}) = 𝒩(𝑥; 𝜇, 𝜎^2)$$

$$(𝜇, \log{𝜎}) = 𝑑_{𝜃}(𝐳)$$

$$\log{𝑝_{𝜃}(𝐱 \mid 𝐳)} = -\frac{1}{2}({\log{2πσ^2}} + \frac{(𝑥 - 𝜇)^2}{σ^2})$$

$$\log{𝑝_{𝜃}(𝐱 \mid 𝐳)} = -\frac{1}{2}({\log{2π}}+\log{σ^2}+\frac{(𝑥 - 𝜇)^2}{σ^2})$$
